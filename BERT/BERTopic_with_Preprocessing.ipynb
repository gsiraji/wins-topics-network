## Load Data

```python
with open('preprocessed_data.json', 'r') as f:
    data = json.load(f)
```

```python
data.keys()
```

```python
cc_abs = data['climate_change_abstract']
cc_ttl = data['climate_change_title']
nlp_abs = data['nlp_abstract']
nlp_ttl = data['nlp_title']
```

## Climate Change

## Abstract

```python
cc_topic_model = BERTopic()
```

```python
cc_topics, cc_probs = cc_topic_model.fit_transform(cc_abs)
```

```python
cc_topic_model.get_topic_info()
```

```python
cc_topic_model.get_topic(0)
```

```python
cc_topic_model.get_document_info(cc_abs)
```

```python
cc_topic_model.visualize_topics()
```
